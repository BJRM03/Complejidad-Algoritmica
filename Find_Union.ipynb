```python
import graphviz as gv

def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p != e:
      graph.edge(str(e), str(p))
  return graph


ds = [0, 1, 2, 2, 8, 5, 5, 7, 8, 2]
print (ds)
drawDS(ds)
```

```python
def find(ds, e):
  return ds[e]

find(ds, 0), find(ds, 7), find(ds, 3), find(ds, 9), find(ds, 2)
```

```python
def union(ds, a, b):
  a = find(ds,  a)
  b = find(ds, b)

  if a == b: return

  for e, p in enumerate(ds):
    if p == a:
      ds[e] = b
```

```python
ds = [i for i in range(10)]
drawDS(ds)
```
