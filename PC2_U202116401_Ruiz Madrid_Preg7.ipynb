```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

aristas = [
    ('D', 'A', 9),
    ('D', 'H', 8),
    ('A', 'E', 5),
    ('A', 'B', 3),
    ('E', 'H', 4),
    ('E', 'I', 4),
    ('E', 'F', 3),
    ('E', 'G', 4),
    ('B', 'F', 1),
    ('F', 'C', 6),
    ('B', 'C', 2),
    ('F', 'G', 3),
    ('F', 'I', 3),
    ('C', 'G', 2),
    ('I', 'G', 7),
    ('H', 'I', 7),
    ('D', 'E', 9)
]

G.add_weighted_edges_from(aristas)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color='blue', edge_color='black', node_size=1000)
etiquetas = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=etiquetas)
plt.title("Grafo Original")
plt.show()

mst = nx.minimum_spanning_tree(G, algorithm='kruskal')

plt.figure(figsize=(10, 6))
nx.draw(mst, pos, with_labels=True, node_color='red', edge_color='black', node_size=1000)
etiquetas_mst = nx.get_edge_attributes(mst, 'weight')
nx.draw_networkx_edge_labels(mst, pos, edge_labels=etiquetas_mst)
plt.title("Grafo Resultante")
plt.show()

peso_total = sum([data['weight'] for u, v, data in mst.edges(data=True)])
print(f"Peso total del Árbol de Expansión Mínimo: {peso_total}")
```
